```python
from typing import TypedDict

class AssistantState(TypedDict):
    message: str
    intent: str
    risk: str
    plan: str
    draft: str
    decision: str
    improve_count: int
    final_answer: str
```

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

OPENROUTER_API_KEY = 'MY_AP_KEY'
OPENROUTER_BASE_URL = 'https://openrouter.ai/api/v1'

# ---لیست مدل‌ها از ضعیف به قوی ) ----
# max_tokens=1024 برای کنترل مصرف اعتبار تعیین شده است
llm_list = [
    ChatOpenAI(   # مدل ۰ - ضعیف‌ترین
        model="google/gemma-3-27b-it",
        temperature=0.2,
        max_tokens=1024,
        api_key=OPENROUTER_API_KEY,
        base_url=OPENROUTER_BASE_URL
    ),
    ChatOpenAI(   # مدل ۱
        model="deepseek/deepseek-v3.2",
        temperature=0.2,
        max_tokens=1024,
        api_key=OPENROUTER_API_KEY,
        base_url=OPENROUTER_BASE_URL
    ),
    ChatOpenAI(   # مدل ۲
        model="qwen/qwen3.6-27b",
        temperature=0.2,
        max_tokens=1024,
        api_key=OPENROUTER_API_KEY,
        base_url=OPENROUTER_BASE_URL
    ),
    ChatOpenAI(   # مدل ۳ - قوی‌ترین
        model="google/gemini-3.5-flash",
        temperature=0.2,
        max_tokens=1024,
        api_key=OPENROUTER_API_KEY,
        base_url=OPENROUTER_BASE_URL
    ),
]

# مدل‌های ثابت برای نودهای اصلی
llm1 = llm_list[1]   # deepseek برای detector، planner، responder
llm2 = llm_list[2]   # qwen برای critic
```

```python
def detect_intent_and_risk(state: AssistantState):
    prompt = f"""
پیام کاربر:
{state["message"]}

۱. هدف پیام را مشخص کن
۲. سطح حساسیت را فقط با یکی از این کلمات تعیین کن:
low
medium
high

خروجی را دقیقاً در این قالب بده:
intent: ...
risk: ...
"""
    result = llm1.invoke(prompt).content
    lines = result.splitlines()
    state["intent"] = lines[0].replace("intent:", "").strip()
    state["risk"] = lines[1].replace("risk:", "").strip()
    return state
```

```python
def planner(state: AssistantState):
    prompt = f"""
کاربر فارسی زبان است.
بر اساس هدف زیر یک برنامه کوتاه برای پاسخ بنویس:

هدف:
{state["intent"]}

سطح حساسیت:
{state["risk"]}
"""
    state["plan"] = llm1.invoke(prompt).content
    return state
```

```python
def responder(state: AssistantState):
    prompt = f"""
بر اساس این برنامه پاسخ اولیه تولید کن:

برنامه:
{state["plan"]}

پیام کاربر:
{state["message"]}
"""
    state["draft"] = llm1.invoke(prompt).content
    return state
```

```python
def critic(state: AssistantState):
    prompt = f"""
وظیفه‌ی تو ارزیابی کیفیت پاسخ داده‌شده به کاربر است.
بر اساس مثال‌های زیر یاد بگیر که چه پاسخ‌هایی مناسب هستند و چه پاسخ‌هایی نیاز به اصلاح دارند.

---
مثال ۱:
درخواست کاربر: آیا می‌توانم بدون وکیل قرارداد ببندم؟
پاسخ داده‌شده: بله می‌توانید.
نتیجه: revise
دلیل: پاسخ بسیار کوتاه و ناقص است و هیچ توضیح یا راهنمایی مفیدی ارائه نمی‌دهد.

---
مثال ۲:
درخواست کاربر: آیا می‌توانم بدون وکیل قرارداد ببندم؟
پاسخ داده‌شده: بله، بستن قرارداد بدون وکیل امکان‌پذیر است، اما ریسک‌هایی دارد. ممکن است بندهای مهمی مانند شرایط فسخ یا ضمانت اجرا از قلم بیفتند. برای قراردادهای ساده می‌توانید از الگوهای استاندارد استفاده کنید، اما برای قراردادهای مهم و پیچیده توصیه می‌شود حتماً با یک وکیل مشورت کنید.
نتیجه: approve
دلیل: پاسخ کامل، دقیق و متناسب با سوال کاربر است و راهنمایی عملی ارائه می‌دهد.

---
مثال ۳:
درخواست کاربر: بهترین روش برای یادگیری برنامه‌نویسی پایتون چیست؟
پاسخ داده‌شده: پایتون یک زبان برنامه‌نویسی است که در سال ۱۹۹۱ ساخته شده است.
نتیجه: revise
دلیل: پاسخ به سوال کاربر درباره روش یادگیری پاسخ نمی‌دهد و فقط اطلاعات تاریخی غیرمرتبط ارائه داده است.

---
مثال ۴:
درخواست کاربر: بهترین روش برای یادگیری برنامه‌نویسی پایتون چیست؟
پاسخ داده‌شده: برای یادگیری پایتون پیشنهاد می‌شود ابتدا با مفاهیم پایه مانند متغیرها، حلقه‌ها و توابع شروع کنید. منابعی مانند مستندات رسمی پایتون، دوره‌های آنلاین مثل Coursera یا یوتیوب و تمرین روزانه با پروژه‌های کوچک بسیار مؤثر هستند. همچنین حل مسائل در سایت‌هایی مثل LeetCode به تقویت مهارت شما کمک می‌کند.
نتیجه: approve
دلیل: پاسخ مستقیماً به سوال کاربر پاسخ داده، منابع عملی معرفی کرده و راهنمایی گام‌به‌گام ارائه داده است.

---
حالا نوبت توست. درخواست کاربر و پاسخ داده‌شده را ارزیابی کن:

درخواست کاربر: {state["message"]}
پاسخ داده‌شده: {state["draft"]}

اگر پاسخ مناسب و کافی است، فقط بنویس:
approve

اگر پاسخ نیاز به اصلاح یا بهبود دارد، فقط بنویس:
revise
"""
    state["decision"] = llm2.invoke(prompt).content.strip()
    # state['critic_count'] += 1
    # if state['critic_count'] > 3:
    #     state["decision"] = "approve"
    return state
```

```python
def improve(state: AssistantState):
    # اگر تمام مدل‌های لیست استفاده شده باشند، از حلقه خارج می‌شویم
    if state["improve_count"] >= len(llm_list):
        state["decision"] = "invalid"
        return state

    # انتخاب مدل بر اساس تعداد دفعات مراجعه به نود improve
    # هر بار از مدل بعدی در لیست (از ضعیف به قوی) استفاده می‌شود
    idx = state["improve_count"]
    current_llm = llm_list[idx]
    state["improve_count"] += 1

    prompt = f"""
این درخواست کاربر است:
{state["message"]}

و این پاسخی است که به او داده شده است:
{state["draft"]}

با توجه به درخواست کاربر، این پاسخ را حرفه‌ای‌تر، دقیق‌تر و امن‌تر کن.
"""
    state["draft"] = current_llm.invoke(prompt).content
    return state
```

```python
def finalize(state: AssistantState):
    state["final_answer"] = state["draft"]
    return state
```

```python
def invalid_request(state: AssistantState):
    # پس از اتمام تمام مدل‌های موجود، درخواست نامعتبر اعلام می‌شود
    state["final_answer"] = (
        "⛔ درخواست نامعتبر است.\n"
        "پس از استفاده از تمام مدل‌های هوش مصنوعی موجود، "
        "هیچ‌کدام نتوانستند پاسخ مناسبی برای این درخواست تولید کنند.\n"
        "لطفاً درخواست خود را بازنویسی کنید یا با پشتیبانی تماس بگیرید."
    )
    return state
```

```python
from langgraph.graph import StateGraph, END

graph = StateGraph(AssistantState)

graph.add_node("detector", detect_intent_and_risk)
graph.add_node("planner", planner)
graph.add_node("responder", responder)
graph.add_node("critic", critic)
graph.add_node("improve", improve)
graph.add_node("finalize", finalize)
graph.add_node("invalid_request", invalid_request)

graph.set_entry_point("detector")

graph.add_edge("detector", "planner")
graph.add_edge("planner", "responder")
graph.add_edge("responder", "critic")

def review_route(state: AssistantState):
    if state["decision"] == "revise":
        return "improve"
    return "finalize"

# مسیریابی از critic: تأیید → finalize، اصلاح → improve
graph.add_conditional_edges(
    "critic",
    review_route,
    {
        "improve": "improve",
        "finalize": "finalize"
    }
)

def improve_route(state: AssistantState):
    # اگر همه مدل‌ها تمام شده‌اند، به نود invalid_request برو
    if state["decision"] == "invalid":
        return "invalid_request"
    return "critic"

# مسیریابی از improve: ادامه چرخه → critic، اتمام مدل‌ها → invalid_request
graph.add_conditional_edges(
    "improve",
    improve_route,
    {
        "critic": "critic",
        "invalid_request": "invalid_request"
    }
)

graph.add_edge("finalize", END)
graph.add_edge("invalid_request", END)

app = graph.compile()
```

```python
result = app.invoke(
    {
        "message": "، چه منابعی را معرفی میکنی ؟ می خواهم در طول یک سال سطح زبان انگلیسی ام حرفه ای شود  ",
        "improve_count": 0
    }
)

print(result["final_answer"])
```
